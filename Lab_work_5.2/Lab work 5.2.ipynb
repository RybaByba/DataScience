Скільки людей користуються англійською мовою "en" на веб-сайті:

```python
def contains(el):
    if 'en' in el.lower():
        return True
    return False

df['Language'].map(contains).sum()
   
```

```python
df[df['Language'].str.contains('en', case=False)] # case=False - без врахування регістру
```

Скільки людей має посаду «Lawyer»?

```python
df[df['Job'].str.contains('Lawyer', case=False)].reset_index()
```
