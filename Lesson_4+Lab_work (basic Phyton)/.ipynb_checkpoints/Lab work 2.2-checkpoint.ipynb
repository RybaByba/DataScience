### Завдання 2

Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16

Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
sex = input('Please indicate your gender. If you`re a women write "f" if you`re a man write "m" >>> ')
age = float(input('Please indicate your age >>> '))

class My_Exception(Exception):
    pass


try:
    
    if sex == 'f' and age >= 18:
        print(f'Your age is, \"{age}\". You can take part in competitions.')

    if sex == 'm' and age >= 16:
        print(f'Your age is, \"{age}\". You can take part in competitions.')
        
    if sex == 'f' and age < 18:
        raise My_Exception({f'"Exception:":"smth wrong", "Message:":"Your age is, {age}. You can not take part in competitions."'})

    if sex == 'm' and age < 16:
        raise My_Exception({f'"Exception:":"smth wrong", "Message:":"Your age is, {age}. You can not take part in competitions."'})
        
except My_Exception as e:
    print("Exception: ", e)

finally:
    print("Program finished")



```

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

# рохраховує мінімальне та максимальне значення
def max_and_min(some_list_оf_1000):
    print(f'max number is:', max(list_with_random_numbers))
    print(f'min number is:', min(list_with_random_numbers))
    
    return max and min



# середнє значення послідовності
def midle(some_list_оf_1000):
    x = sum(list_with_random_numbers)/1000
    print(f'average value: ', round(x, 2))
    
    return x


# медіану послідовності  
def mediana_by_list(some_list_оf_1000):
    list_with_random_numbers.sort()
    elem1 = len(list_with_random_numbers)//2
    elem2 = len(list_with_random_numbers)//2-1
    mediana = (list_with_random_numbers[elem1] + list_with_random_numbers[elem2])/2
    print(f'mediana for 1000 numbers:', mediana)
    
    return mediana


 
list_numbers = [1,45,15,18,20]

def mediana_by_any_list(list_numbers):
    list_numbers.sort()
    
    if len(list_numbers) % 2 == 0:
        elem1 = len(list_numbers)//2
        elem2 = len(list_numbers)//2-1
        mediana = (list_numbers[elem1] + list_numbers[elem2])/2
        
    else:
        mediana = len(list_numbers)/2
    print(f'mediana for any numbers:', mediana)
    
    return mediana


# частоту кожної цифри в послідовності
def dict_number(some_list):
    number_dict = {}
    for number in some_list:
        if number in number_dict:
            number_dict[number] = number_dict[number] + 1
        else:
            number_dict[number] = 1
    return number_dict



# середнє квадратичне відхилення
from math import sqrt

def standard_deviation(some_list):
    count = 0

    for i in some_list:
        count += i
#     print(f'sum all numbers: {count}')
    md = count // len(some_list)
#     print(f'average value all numbers: {md}')
    count_1 = 0

    for i in some_list:
        count_1 += (i - md)**2
#     print(f'squared deviation: {count_1}')
    dispersion = count_1 // len(some_list)
#     print(f'dispersion: {dispersion}')
    standard_deviation = float(sqrt(dispersion))
#     print(f'standard_deviation: {standard_deviation}')
    
    return standard_deviation



max_and_min(list_with_random_numbers)
midle(list_with_random_numbers)
mediana_by_list(list_with_random_numbers)
mediana_by_any_list(list_numbers)

print(f'''
The frequency of each numbers in the list:

{dict_number(list_with_random_numbers)}

standard_deviation: {round(standard_deviation(list_with_random_numbers), 2)}
''')
```
